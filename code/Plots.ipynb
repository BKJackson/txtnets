```python
with open("../data/stanfordmovie/stanfordmovie.train.sentences.clean.json") as train_file:
    train_data = json.load(train_file)
```

```python
doc_lengths = []
sentence_lengths = []
for doc, label in train_data:
    doc_lengths.append(len(doc))
    sentence_lengths.extend(len(s) for s in doc)
```

```python
plt.figure(figsize=(20, 10))
plt.hist(doc_lengths, 100);
plt.xlim([0, 75])
len(filter(lambda x: x <= 30, doc_lengths)) / float(len(doc_lengths))
```

```python
plt.hist(sentence_lengths, 100);
plt.xlim([0, 100])
len(filter(lambda x: x <= 50, sentence_lengths)) / float(len(sentence_lengths))
```
