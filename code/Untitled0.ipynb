```python
params = [
    'n_feature_maps_w_1', 'n_feature_maps_w_2', 'n_feature_maps_s_1',
    'kernel_width_w_1', 'kernel_width_w_2', 'kernel_width_s_1',
    'k_pooling_w_1', 'k_pooling_w_2', 'k_pooling_s_1',
    'dropout',
    ]

with open("../results/pandas_data_jasper_basic_model_randomsearch_2014-05-31_2.pkl") as data_file:
    df = pd.DataFrame(pickle.load(data_file))
    df = df[df['best_validation_accuracy'] > 0.7]

for param in params:
    subplot_groups = df.groupby(param)
    n_subplots = len(subplot_groups)
    fig, axes = plt.subplots(1, len(subplot_groups), figsize=(4*n_subplots, 4))
    try:
        iter(axes)
    except:
        axes = [axes]
    for ax, (name, group) in zip(axes, subplot_groups):
        ax.set_title("{}={}".format(param, name))
        line_groups = group.groupby('job_id')
        for line_name, line_group in line_groups:
            ax.plot(line_group['batch'], line_group['validation_accuracy'], label=line_name)
```
