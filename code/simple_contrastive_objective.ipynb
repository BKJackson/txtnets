```python
%pylab inline --no-import-all

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1232)
```

```python
class Linear(object):
    def __init__(self, n_in, n_out):
        self.n_in = n_in
        self.n_out = n_out
        self.w = 0.1 * np.random.standard_normal(size=(n_in, n_out))
        
#         k = min(n_in, n_out)
#         u, _, v = np.linalg.svd(self.w)
#         self.w = np.dot(u[:,:k], v[:k,:])
        
        self.b = np.zeros(shape=(1, n_out))

    def fprop(self, x):
        return np.dot(x, self.w) + self.b

    def bprop(self, x, delta):
        return np.dot(delta, self.w.T)

    def grads(self, x, delta):
        return [np.dot(x.T, delta), np.sum(delta, axis=0, keepdims=True)]
```

```python
class GaussianEnergy(object):
    def fprop(self, x, y):
        return 0.5 * np.sum((x - y)**2, axis=1, keepdims=True)

    def bprop(self, x, y, delta):
        delta_x = delta * x
        delta_y = delta * y
        return delta_x, delta_y
```

```python
class ContrastiveHingeLoss(object):
    def __init__(self, margin):
        self.margin = margin
        
    def fprop(self, x_clean, x_dirty):
        return np.maximum(0.0, self.margin + x_clean - x_dirty)

    def bprop(self, x_clean, x_dirty):
        delta = self.fprop(x_clean, x_dirty) > 0
        delta = delta.astype(np.float)
        delta /= float(x_clean.shape[0])
        delta_clean = delta * x_clean
        delta_dirty = delta * -x_dirty
        return delta_clean, delta_dirty
```

```python
class LatentBilingualModel(object):
    def __init__(self, n_latent, n_visible):
        self.n_latent = n_latent
        self.n_visible = n_visible
        
        self.w1 = np.random.standard_normal(size=(n_latent, n_visible))
        self.w2 = np.random.standard_normal(size=(n_latent, n_visible))
        
        k = min(n_latent, n_visible)
        
        u, _, v = np.linalg.svd(self.w1)
        self.w1 = np.dot(u[:,:k], v[:k,:])
        self.w1 = np.eye(*self.w1.shape)
        
        u, _, v = np.linalg.svd(self.w2)
        self.w2 = np.dot(u[:,:k], v[:k,:])
        
        self.w2 = self.w1
        
        self.b1 = 2 * np.random.standard_normal(size=(1, n_latent))
        self.b2 = 2 * np.random.standard_normal(size=(1, n_latent))
        
    def generate(self, n):
        # z = np.random.standard_normal(size=(n, self.n_latent))
        z = np.random.uniform(-10, 10, size=(n, self.n_latent))
        return z
    
    def observe(self, z):
        x1 = np.dot(z - self.b1, self.w1)
        x2 = np.dot(z - self.b2, self.w2)
        return x1, x2
```

```python
d = 2

generative_model = LatentBilingualModel(n_latent=2, n_visible=d)
z = generative_model.generate(100)
x1_clean, x2_clean = generative_model.observe(z)

model1 = Linear(n_in=d, n_out=2)
model2 = Linear(n_in=d, n_out=2)
model1.w = generative_model.w1.copy()
model2.w = model1.w.copy()
# model2 = model1

fig = plt.figure()
ax = fig.gca()
y1 = model1.fprop(x1_clean)
y2 = model2.fprop(x2_clean)
ax.scatter(y1[:,0], y1[:,1], c='red');
ax.scatter(y2[:,0], y2[:,1], c='blue');
```

```python
energy = GaussianEnergy()
loss = ContrastiveHingeLoss(margin=1.0)

loss_values = []
energy_clean_values = []
energy_noise_values = []

for epoch in xrange(5000):
    y1_clean = model1.fprop(x1_clean)
    y2_clean = model2.fprop(x2_clean)
    
    x1_noise = np.random.permutation(x1_clean)
    y1_noise = model1.fprop(x1_noise)

    x2_noise = x2_clean #np.random.permutation(x2_clean)
    y2_noise = model2.fprop(x2_noise)

    e_clean = energy.fprop(y1_clean, y2_clean)
    e_noise = energy.fprop(y1_noise, y2_noise)

    l = loss.fprop(e_clean, e_noise)
    loss_values.append(l.mean())
    
    energy_clean_values.append(e_clean.mean())
    energy_noise_values.append(e_noise.mean())

    dloss_clean, dloss_noise = loss.bprop(e_clean, e_noise)

    denergy_1_clean, denergy_2_clean = energy.bprop(y1_clean, y2_clean, dloss_clean)
    denergy_1_noise, denergy_2_noise = energy.bprop(y1_noise, y2_noise, dloss_noise)

    gw_1_clean, gb_1_clean = model1.grads(x1_clean, denergy_1_clean)
    gw_1_noise, gb_1_noise = model1.grads(x1_noise, denergy_1_noise)
    
    gw_2_clean, gb_2_clean = model2.grads(x2_clean, denergy_2_clean)
    gw_2_noise, gb_2_noise = model2.grads(x2_noise, denergy_2_noise)

    gw_1 = gw_1_clean + gw_1_noise
    gw_2 = gw_2_clean + gw_2_noise
    
    gb_1 = gb_1_clean + gb_1_noise
    gb_2 = gb_2_clean + gb_2_noise
    
    model1.w -= 0.005 / np.sqrt(epoch + 1.0) * gw_1
    model2.w -= 0.005 / np.sqrt(epoch + 1.0) * gw_2
    
    model1.b -= 0.005 / np.sqrt(epoch + 1.0) * gb_1
    model2.b -= 0.005 / np.sqrt(epoch + 1.0) * gb_2
    
#     loss.margin *= 1.0005
    
# loss_values = pd.Series(data=loss_values, name="loss")
plt.figure();
sns.tsplot(loss_values);
plt.figure();
sns.tsplot(energy_clean_values);
plt.figure();
sns.tsplot(energy_noise_values);
```

```python
fig = plt.figure()
ax = fig.gca()
y1 = model1.fprop(x1_clean)
y2 = model2.fprop(x2_clean)
ax.scatter(y1[:,0], y1[:,1], c='red');
ax.scatter(y2[:,0], y2[:,1], c='blue');
ax.set_title("Embeddings");
```
