I ran experiments testing different batch sizes, different learning rates and different numbers of negative samples.

A learning rate of 0.1 is too large, the objective value typically remains flat or occasionaly has large jumps (in the wrong direction).

With one negative sample nothing is learned.  This suggests I should try the alternative formulation of this with the siamese net objective.

With 50 negative samples and a small learning rate the objective value decreases with both batch sizes (50 and 100).  Using smaller batches is slower unfortunately.

```python
%pylab inline --no-import-all

import numpy as np
import pandas as pd
import seaborn as sns
import re
import os
import pstats
import glob

HERE = os.environ['HERE']
```

```python
train_info = pd.DataFrame()
files = glob.glob(os.path.join(os.environ['RESULTS'], "run_2014-05-23_10-28-32/*.log"))

column_names = {
    "B": "batch",
    "E": "epoch",
    "L": "loss",
    "C": "correct dist2",
    "I": "incorrect dist2",
    "I-C": "ic2-c2",
    "EPH": "examples/hr",
    "PS": "param size",
}

for f in files:
    with open(f) as train_log_file:
        for line in train_log_file:
            matches = re.findall("([^ :]+): ([0-9.]+)", line)
            if matches and re.match("B:", line):
                cols = dict((column_names[name_code], float(value)) for name_code, value in matches)
                cols['file'] = f
                train_info = train_info.append(cols, ignore_index=True)
```
